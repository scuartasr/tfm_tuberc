```python
# Imports y configuración de paths
import os, sys, numpy as np, pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

PROJ_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..', '..'))
DATA_PATH = os.path.join(PROJ_ROOT, 'data', 'processed', 'mortalidad')
BASE_OUT = os.path.join(PROJ_ROOT, 'modelos', 'outputs')
NOTEBOOK_NAME = os.path.splitext(os.path.basename(os.environ.get('NOTEBOOK_BASENAME', 'b_07_ml.ipynb')))[0]
OUT_DIR = os.path.join(BASE_OUT, 'output', NOTEBOOK_NAME)
os.makedirs(OUT_DIR, exist_ok=True)

print(f'Datos: {DATA_PATH}')
print(f'Salidas: {OUT_DIR}')

pd.set_option('display.width', 140)
pd.set_option('display.max_columns', 60)
sns.set_context('talk')
plt.rcParams['figure.figsize'] = (10,5)
```

```python
# Carga de datos y matriz edad×año (m y log(m))
EPS = 1e-12
file_long = os.path.join(DATA_PATH, 'tasas_mortalidad_gret_per.csv')
df_long = pd.read_csv(file_long).rename(columns={'ano':'year','gr_et':'age'})
df_long_m = df_long[['year','age','tasa_x100k']].copy()
df_long_m['m'] = (df_long_m['tasa_x100k'] / 1e5).clip(lower=EPS)
mat = df_long_m.pivot(index='age', columns='year', values='m').sort_index()
ages = mat.index.values.astype(int)
years = mat.columns.values.astype(int)
mat_log = np.log(mat.clip(lower=EPS))
print('Matriz m:', mat.shape, '| años', years.min(), '-', years.max(), '| edades', ages.min(), '-', ages.max())
mat.head()
```

```python
# Construcción de secuencias y split temporal
T = 10  # ventanas de historial
cut_year = 2018
idx_cut = int(np.where(years == cut_year)[0][0])
# Normalización (solo con train) en escala log
train_logs = mat_log.iloc[:, :idx_cut+1].values.reshape(-1)
mu, sigma = float(np.mean(train_logs)), float(np.std(train_logs) + 1e-8)

# Generar secuencias train para todas las edades
X_list, y_list, meta_age, meta_year = [], [], [], []
for age in ages:
    s = mat_log.loc[age].values
    for t in range(T, idx_cut+1):  # predice años <= 2018
        X_list.append((s[t-T:t] - mu)/sigma)
        y_list.append((s[t] - mu)/sigma)
        meta_age.append(int(age))
        meta_year.append(int(years[t]))
X_train = np.array(X_list, dtype=np.float32)[..., np.newaxis]  # (N,T,1)
y_train = np.array(y_list, dtype=np.float32)[..., np.newaxis]  # (N,1)
print('Secuencias train:', X_train.shape, y_train.shape)
```

```python
# Búsqueda aleatoria de hiperparámetros para LSTM
import json, random, time
np.random.seed(42)
try:
    tf.random.set_seed(42)
except Exception:
    pass

# Espacio de búsqueda (puede ajustarse por variable de entorno)
search_space = {
    'units': [16, 32, 64, 96],
    'stacked': [False, True],
    'dropout': [0.0, 0.2, 0.4],
    'lr': [1e-3, 5e-4, 1e-4],
    'batch_size': [16, 32, 64]
}

n_trials = int(os.environ.get('HP_N_TRIALS', 15))
max_epochs = int(os.environ.get('HP_MAX_EPOCHS', 120))
patience = int(os.environ.get('HP_PATIENCE', 15))
val_split = 0.1

from tensorflow.keras import layers, models, callbacks, optimizers


def build_model(hp: dict):
    m = models.Sequential()
    m.add(layers.Input(shape=(T, 1)))
    if hp['stacked']:
        m.add(layers.LSTM(hp['units'], activation='tanh', return_sequences=True))
        if hp['dropout'] > 0:
            m.add(layers.Dropout(hp['dropout']))
        m.add(layers.LSTM(max(8, hp['units'] // 2), activation='tanh'))
    else:
        m.add(layers.LSTM(hp['units'], activation='tanh'))
    if hp['dropout'] > 0:
        m.add(layers.Dropout(hp['dropout']))
    m.add(layers.Dense(1))
    opt = optimizers.Adam(learning_rate=hp['lr'])
    m.compile(optimizer=opt, loss='mse')
    return m


def sample_hparams():
    return {k: random.choice(v) for k, v in search_space.items()}

trials = []
best = None
start_all = time.time()
for i in range(1, n_trials + 1):
    hp = sample_hparams()
    tf.keras.backend.clear_session()
    model_i = build_model(hp)
    es = callbacks.EarlyStopping(monitor='val_loss', patience=patience, restore_best_weights=True, verbose=0)
    t0 = time.time()
    hist = model_i.fit(
        X_train, y_train,
        validation_split=val_split,
        epochs=max_epochs,
        batch_size=hp['batch_size'],
        verbose=0,
        callbacks=[es]
    )
    dur = time.time() - t0
    val_min = float(np.min(hist.history['val_loss']))
    epochs_run = len(hist.history['loss'])
    rec = dict(trial=i, val_loss=val_min, epochs=epochs_run, seconds=round(dur, 2), **hp)
    trials.append(rec)
    if (best is None) or (val_min < best['val_loss']):
        best = rec
    print(f"Trial {i:02d}/{n_trials} -> val_loss={val_min:.5f} | {hp} | epochs={epochs_run} | {dur:.1f}s")

print('Mejor configuración:', best)

# Persistir resultados
results_df = pd.DataFrame(trials).sort_values('val_loss').reset_index(drop=True)
results_path = os.path.join(OUT_DIR, 'lstm_hparam_trials.csv')
results_df.to_csv(results_path, index=False)
with open(os.path.join(OUT_DIR, 'lstm_best_hparams.json'), 'w') as f:
    json.dump(best, f, indent=2)

# Reentrenar modelo final con mejores hiperparámetros
tf.keras.backend.clear_session()
best_hparams = {k: best[k] for k in ['units', 'stacked', 'dropout', 'lr', 'batch_size']}
model = build_model(best_hparams)
es_final = callbacks.EarlyStopping(monitor='val_loss', patience=patience, restore_best_weights=True)
hist = model.fit(
    X_train, y_train,
    validation_split=val_split,
    epochs=max_epochs,
    batch_size=best_hparams['batch_size'],
    verbose=0,
    callbacks=[es_final]
)
print('Entrenamiento final completado con hiperparámetros óptimos. Épocas:', len(hist.history['loss']))
print('Resultados guardados en:', OUT_DIR)
```

# Optimización de hiperparámetros
En esta sección realizamos una búsqueda aleatoria de hiperparámetros para el modelo LSTM usando `EarlyStopping` sobre un `validation_split` del 10%. Guardamos todas las corridas en `lstm_hparam_trials.csv` y la mejor configuración en `lstm_best_hparams.json` dentro de `OUT_DIR`. Finalmente reentrenamos el modelo con la mejor configuración y lo dejamos en la variable `model`.

```python
# Definición y entrenamiento del modelo LSTM
from tensorflow.keras import layers, models, callbacks, optimizers

# Si ya existe un modelo entrenado por tuning, omitir este bloque
_has_best = 'best_hparams' in globals() and isinstance(globals().get('best_hparams'), dict)
if _has_best:
    print('Modelo ya entrenado con hiperparámetros óptimos. Omitiendo entrenamiento básico.')
else:
    model = models.Sequential([
        layers.Input(shape=(T, 1)),
        layers.LSTM(32, activation='tanh'),
        layers.Dense(1)
    ])

    opt = optimizers.Adam(learning_rate=1e-3)
    model.compile(optimizer=opt, loss='mse')

    es = callbacks.EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
    hist = model.fit(X_train, y_train, validation_split=0.1, epochs=200, batch_size=32, verbose=0, callbacks=[es])

    print('Entrenamiento completado. Épocas:', len(hist.history['loss']))
```

```python
# Predicciones one-step para train (1979–2018)
yhat_train_scaled = model.predict(X_train, verbose=0).reshape(-1)
yhat_train_log = (yhat_train_scaled * sigma) + mu
ytrue_train_log = np.array([mat_log.loc[a, y] for a, y in zip(meta_age, meta_year)])
m_train_pred = np.exp(yhat_train_log)
m_train_obs = np.exp(ytrue_train_log)
train_df = pd.DataFrame({'age': meta_age, 'year': meta_year, 'm_obs': m_train_obs, 'm_pred': m_train_pred})
print('Predicciones train:', train_df.shape)
```

```python
# Predicciones recursivas para validación (2019–2023)
val_years = years[idx_cut+1:]
Xv_meta_age, Xv_meta_year, yhat_val_log = [], [], []
for age in ages:
    s = mat_log.loc[age].values.copy()
    context = s[idx_cut+1-T:idx_cut+1].copy()  # últimos T hasta 2018
    for yy in val_years:
        xin = ((context - mu)/sigma).astype('float32').reshape(1, T, 1)
        y_scaled = float(model.predict(xin, verbose=0).reshape(-1)[0])
        y_log = y_scaled * sigma + mu
        yhat_val_log.append(y_log)
        Xv_meta_age.append(int(age)); Xv_meta_year.append(int(yy))
        context = np.concatenate([context[1:], [y_log]])

m_val_pred = np.exp(np.array(yhat_val_log))
m_val_obs = np.array([mat.loc[a, y] for a, y in zip(Xv_meta_age, Xv_meta_year)])
val_df = pd.DataFrame({'age': Xv_meta_age, 'year': Xv_meta_year, 'm_obs': m_val_obs, 'm_pred': m_val_pred})
print('Predicciones validación:', val_df.shape)
```

```python
# Métricas (m y log)
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

def smape(y_true, y_pred, eps=1e-12):
    return 100 * np.mean(2*np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true) + eps))

def wape(y_true, y_pred, eps=1e-12):
    return 100 * (np.sum(np.abs(y_pred - y_true)) / (np.sum(np.abs(y_true)) + eps))

# Train
RMSE_train = float(np.sqrt(mean_squared_error(train_df['m_obs'], train_df['m_pred'])))
MAE_train = float(mean_absolute_error(train_df['m_obs'], train_df['m_pred']))
sMAPE_train = float(smape(train_df['m_obs'], train_df['m_pred']))
WAPE_train = float(wape(train_df['m_obs'], train_df['m_pred']))
RMSE_log_train = float(np.sqrt(mean_squared_error(np.log(train_df['m_obs'].clip(lower=EPS)), yhat_train_log)))
MSE_train = float(mean_squared_error(train_df['m_obs'], train_df['m_pred']))
MAPE_train = 100 * float(mean_absolute_percentage_error(train_df['m_obs'], train_df['m_pred']))
# Calcular MAPE y MSE


# Val
RMSE_val = float(np.sqrt(mean_squared_error(val_df['m_obs'], val_df['m_pred'])))
MSE_val = float(mean_squared_error(val_df['m_obs'], val_df['m_pred']))
MAPE_val = 100 * float(mean_absolute_percentage_error(val_df['m_obs'], val_df['m_pred']))
MAE_val = float(mean_absolute_error(val_df['m_obs'], val_df['m_pred']))
sMAPE_val = float(smape(val_df['m_obs'], val_df['m_pred']))
WAPE_val = float(wape(val_df['m_obs'], val_df['m_pred']))
RMSE_log_val = float(np.sqrt(mean_squared_error(np.log(val_df['m_obs'].clip(lower=EPS)), np.array(yhat_val_log))))

print('Métricas (escala original por persona):')
print(f'- Train | RMSE: {RMSE_train:.4e} | MSE: {MSE_train:.4e} | MAPE: {MAPE_train:.2e}% | MAE: {MAE_train:.4e} | sMAPE: {sMAPE_train:.2f}% | WAPE: {WAPE_train:.2f}%')
print(f'- Val   | RMSE: {RMSE_val:.4e} | MSE: {MSE_val:.4e} | MAPE: {MAPE_val:.2e}% | MAE: {MAE_val:.4e} | sMAPE: {sMAPE_val:.2f}% | WAPE: {WAPE_val:.2f}%')

print('Métricas (escala log):')
print(f'- Train | RMSE_log: {RMSE_log_train:.4e}')
print(f'- Val   | RMSE_log: {RMSE_log_val:.4e}')
```

```python
# Cálculo de valores SHAP (KernelExplainer)
try:
    import shap
except Exception:
    import subprocess, sys as _sys
    subprocess.check_call([_sys.executable, '-m', 'pip', 'install', '-q', 'shap'])
    import shap

# Validaciones previas
if 'model' not in globals():
    raise RuntimeError('El modelo no está definido. Ejecuta las celdas de entrenamiento primero.')
if 'X_train' not in globals():
    raise RuntimeError('X_train no está disponible. Ejecuta la preparación de datos.')

# Preparar funciones y subconjuntos
bg_size = int(os.environ.get('SHAP_BG_SIZE', 128))  # más conservador por tiempo
sample_size = int(os.environ.get('SHAP_SAMPLE_SIZE', 256))

# Predicción 1D para KernelExplainer
def f_pred(x):
    # x como (N, T, 1) -> salida (N,)
    y = model.predict(x, verbose=0).reshape(-1)
    return y

# Selección de background y muestra
total_n = X_train.shape[0]
bg_n = min(bg_size, total_n)
sm_n = min(sample_size, total_n)
idx_bg = np.random.choice(total_n, size=bg_n, replace=False)
idx_sm = np.random.choice(total_n, size=sm_n, replace=False)
X_bg = X_train[idx_bg]
X_sm = X_train[idx_sm]

# Construir explainer y calcular SHAP
explainer = shap.KernelExplainer(f_pred, X_bg)
# nsamples='auto' puede ser costoso; limitar por tiempo
shap_values = explainer.shap_values(X_sm, nsamples=max(100, T * 20))
if isinstance(shap_values, list):
    shap_values = shap_values[0]

# Guardar valores SHAP
np.save(os.path.join(OUT_DIR, 'shap_values.npy'), shap_values)

# Importancia por lag (media de |SHAP| a lo largo de ejemplos)
abs_mean = np.mean(np.abs(shap_values), axis=0).reshape(-1)  # (T,)
lag_idx = np.arange(T)
imp_df = pd.DataFrame({'lag_index': lag_idx, 'abs_mean_shap': abs_mean})
imp_df.to_csv(os.path.join(OUT_DIR, 'shap_feature_importance.csv'), index=False)

# Graficar importancia por lag
plt.figure(figsize=(9,4))
plt.bar(lag_idx, abs_mean, color=get_palette(3)[1])
plt.xticks(lag_idx, [f't-{T-i}' for i in range(1, T+1)], rotation=0)
plt.xlabel('Lag temporal en la ventana')
plt.ylabel('Importancia media |SHAP|')
plt.title('Importancia de lags para LSTM (SHAP KernelExplainer)')
plt.grid(axis='y', alpha=0.3)
out_imp = os.path.join(OUT_DIR, 'lstm_shap_importance_lags.png')
plt.tight_layout(); plt.savefig(out_imp, dpi=160); plt.show()
print('SHAP listo. Archivos en:', OUT_DIR)
print('Muestra usada | background:', X_bg.shape, '| sample:', X_sm.shape)
```
